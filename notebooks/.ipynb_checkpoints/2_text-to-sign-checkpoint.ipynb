```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

import spacy
nlp = spacy.load("en_core_web_sm")
print("spaCy loaded!")
```

```python
def build_clip_index(clips_dir):
    clip_index = {}
    for root, dirs, files in os.walk(clips_dir):
        for clip in files:
            if not clip.lower().endswith(".mp4"):
                continue
            clip_name = clip.replace(".mp4", "")
            clip_name = clip_name.replace(" - English", "")
            clip_name = clip_name.replace(" - Hindi", "")
            clip_name = clip_name.replace("_", " ")
            clip_name = clip_name.lower().strip()

            clip_index[clip_name] = os.path.join(root, clip)
    return clip_index
```

```python
BASE_DIR = os.path.dirname(os.path.abspath("__file__"))  #get the absolute path of the current file's directory

SETU_DIR = os.path.dirname(BASE_DIR) #move one level up to reach the project root directory

CLIPS_DIR = os.path.join(SETU_DIR, "data", "isl_clips") #define the path to the dataset containing isl clips
CLIP_INDEX = build_clip_index(CLIPS_DIR)

MODELS_DIR = os.path.join(SETU_DIR, "models") #define the path to the directory where trained models are stored

# verify
print("CLIPS DIR:", CLIPS_DIR)
print("MODELS DIR:", MODELS_DIR)

print("indexed clips:", len(CLIP_INDEX))
```

```python
# ── Word Mappings ──
WORD_MAPPINGS = {
    'i': 'i me',
    'me': 'i me',
    'thank': 'thank you',
    'thankyou': 'thank you',
    'toilet': 'toile',
    'need': 'want desire wish for',
    'want': 'want desire wish for',
    'how': 'how (explanation)',
    'who': 'who whom',
    'why': 'why (explanation)',
    'when': 'when (for days)',
}

# ── ISL Grammar Word Lists ──
TIME_WORDS = ['morning', 'evening', 'night', 'today', 'yesterday',
              'tomorrow', 'now', 'later', 'always', 'soon']

TIME_MODIFIERS = ['good']

QUESTION_WORDS = ['what', 'where', 'when', 'who', 'how', 'why', 'which']

NEGATION_WORDS = ['not', 'no', "don't", "doesn't", "won't",
                  "can't", "isn't", "aren't", "didn't", "n't"]

ALWAYS_KEEP = ['hello', 'goodbye', 'yes', 'no', 'sorry', 'please',
               'i', 'you', 'we', 'they', 'he', 'she', 'your', 'my']

OBJECT_PRONOUNS = ['me', 'him', 'her', 'them', 'us']

ALWAYS_REMOVE = ['feeling', 'having', 'getting', 'doing',
                 'coming', 'being']

FORCE_KEEP = ['name', 'place', 'age', 'time', 'day']

```

```python
def find_clip(word):
    word_lower = word.lower().strip()

    if word_lower in WORD_MAPPINGS:
        word_lower = WORD_MAPPINGS[word_lower].lower()

    # Exact match
    if word_lower in CLIP_INDEX:
        return CLIP_INDEX[word_lower]

    # Partial match
    for key, path in CLIP_INDEX.items():
        if word_lower in key:
            return path

    return None
```

```python
def play_clip(clip_path):
    cap = cv2.VideoCapture(clip_path)  # open video file
    while cap.isOpened():
        ret, frame = cap.read()  # read frame
        if not ret:
            break  # video ended
        cv2.imshow('SETU - ISL Sign', frame)  # show frame
        cv2.waitKey(33)  # wait 33ms between frames = 30fps
    cap.release()
    cv2.destroyAllWindows()

```

```python
def text_to_isl_words(sentence, nlp):
    doc = nlp(sentence.lower()) #this gives information like which parts of speech is .
    
    time_bucket = []
    pronoun_bucket = []
    noun_bucket = []
    verb_bucket = []
    object_bucket = []   # me, him, her — after verbs
    negation_bucket = []
    question_bucket = []
    other_bucket = []
    
    for token in doc:
        word = token.lemma_.lower()

        if word == "-pron-":
            word = token.text.lower()
        
        # skip always remove
        if word in ALWAYS_REMOVE:
            continue
        
        # negation handling like 'i do not like coffee' to 'i coffee like no'
        if word in NEGATION_WORDS or token.lemma_ in ['not', 'never']:
            if 'no' not in negation_bucket:
                negation_bucket.append('no')
        
        # time modifiers + time words go together in time bucket
        elif word in TIME_MODIFIERS:
            time_bucket.append(word)
        
        elif word in TIME_WORDS:
            time_bucket.append(word)
        
        # question words last
        elif word in QUESTION_WORDS:
            question_bucket.append(word)
        
        # object pronouns after verbs
        elif word in OBJECT_PRONOUNS:
            object_bucket.append(word)
        
        # subject pronouns
        elif word in ALWAYS_KEEP:
            if token.pos_ == 'PRON' or word in ['i', 'you', 'we', 
                                                  'they', 'he', 'she']:
                pronoun_bucket.append(word)
            else:
                other_bucket.append(word)
        
        # force keep important words spaCy filters
        elif word in FORCE_KEEP:
            noun_bucket.append(word)
        
        # skip remaining stop words
        elif token.is_stop:
            continue
        
        # nouns and adjectives
        elif token.pos_ in ['NOUN', 'PROPN', 'ADJ', 'NUM']:
            noun_bucket.append(word)
        
        # verbs
        elif token.pos_ == 'VERB':
            verb_bucket.append(word)
    
    # ISL order: TIME → GREETINGS → PRONOUN → NOUN → VERB → OBJECT → NEGATION → QUESTION
    result = (time_bucket + other_bucket + pronoun_bucket +
              noun_bucket + verb_bucket + object_bucket +
              negation_bucket + question_bucket)
    
    return [w.upper() for w in result]

```

```python
def is_idle_frame(frame):
    # convert to grayscale and check if frame is mostly static/blank
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # if standard deviation is very low = static frame
    return gray.std() < 15
```

```python
def text_to_sign_smooth(sentence, nlp, clips_dir):
    isl_words = text_to_isl_words(sentence, nlp)
    print(f"Input: {sentence}")
    print(f"ISL words: {isl_words}")
    
    for word in isl_words:
        clip_path = find_clip(word)
        if clip_path:
            cap = cv2.VideoCapture(clip_path)
            total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
            
            frame_num = 0
            while cap.isOpened():
                ret, frame = cap.read()
                if not ret:
                    break
                
                frame = cv2.resize(frame, (640, 480),
                                 interpolation=cv2.INTER_AREA)
                
                # overlay the current word at the top left corner
                cv2.putText(frame, word, (10, 50),
                           cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0,255,0), 3)
                
                # progress bar at bottom
                progress = int((frame_num / total_frames) * 640)
                cv2.rectangle(frame, (0, 465), (progress, 480), (0,255,0), -1)
                
                cv2.imshow('SETU - ISL Signs', frame)
                cv2.waitKey(40)
                frame_num += 1
            
            cap.release()
        else:
            blank = np.zeros((480, 640, 3), dtype=np.uint8)
            cv2.putText(blank, f"No sign: {word}", (50, 240),
                       cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2)
            cv2.imshow('SETU - ISL Signs', blank)
            cv2.waitKey(800)  # show for 0.8 seconds then continue
    
    cv2.destroyAllWindows()

```

```python
# Coverage Test

test_sentences = [
    "I need water",
    "The doctor is helping the students",
    "Go to the hospital",
    "The teacher is teaching",
    "Thank you very much",
    "Good morning",
    "Students are learning",
    "Doctors help patients",
    "Where is the railway station",
    "I want to learn python",
    "My friend is studying",
    "The engineer is working",
    "Please help me",
    "I am hungry",
    "The child is playing",
    "WHERE are you going",
    "should you go home now",
    "tomorrow we are going away",
    "what shall we do",
    "what ever it is let it be"
]

total_words = 0
found_words = 0
missing_words = set()

for sentence in test_sentences:

    isl_words = text_to_isl_words(sentence, nlp)

    print(f"\nSentence: {sentence}")
    print("ISL:", isl_words)

    for word in isl_words:

        total_words += 1

        clip = find_clip(word.lower())

        if clip:
            found_words += 1
            print(f"✓ {word}")
        else:
            missing_words.add(word)
            print(f"✗ {word}")

coverage = (found_words / total_words) * 100

print("\n" + "=" * 50)
print("COVERAGE TEST RESULTS")
print("=" * 50)

print(f"Total ISL words: {total_words}")
print(f"Found clips: {found_words}")
print(f"Missing clips: {total_words - found_words}")
print(f"Coverage: {coverage:.2f}%")

print("\nMissing Words:")
for word in sorted(missing_words):
    print(word)
```

```python
import cv2

input_path = r"C:\Users\thiru\SETU\data\isl_clips\H\How_(Explanation)_backup.mp4"
output_path = r"C:\Users\thiru\SETU\data\isl_clips\H\How_(Explanation).mp4"

cap = cv2.VideoCapture(input_path)
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

keep_frames = int(fps * 4)  # 4 seconds

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

frame_count = 0
while True:
    ret, frame = cap.read()
    if not ret or frame_count >= keep_frames:
        break
    out.write(frame)
    frame_count += 1

cap.release()
out.release()
print(f"Done! Saved {frame_count/fps:.1f} second clip")
```
